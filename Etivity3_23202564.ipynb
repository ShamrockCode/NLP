```python
import random

def ShannonVisualization(bigram_counts, seed="<s>"):
    sentence = [seed]

    while True:
        # Retrieve all bigrams starting with the current seed word
        candidates = [(w, c) for (s, w), c in bigram_counts.items() if s == seed]

        # If no bigrams found or the seed word indicates the end of a sentence, break
        if not candidates or seed == "</s>":
            break

        # Compute the cumulative probability intervals for all candidates
        total = sum([c for _, c in candidates])
        cum_prob = 0
        rand_num = random.random()

        for word, count in candidates:
            prob = count / total
            if cum_prob <= rand_num < cum_prob + prob:
                sentence.append(word)
                seed = word
                break
            cum_prob += prob

    return ' '.join(sentence)


bigram_counts = {
    ("<s>", "I"): 5,
    ("I", "want"): 3,
    ("want", "to"): 2,
    ("to", "eat"): 1,
    ("eat", "Chinese"): 1,
    ("Chinese", "food"): 1,
    ("food", "</s>"): 1
}

# Generate a sentence
print(ShannonVisualization(bigram_counts))
```
