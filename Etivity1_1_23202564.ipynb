```python
import re
import csv

!wget https://gist.githubusercontent.com/ajoorabchi/eac194a79dd26de8864f9206b7842ff1/raw/8ea1d8d5f74b5b2724e378b43d4df6094990c7db/Eircode%2520Routing%2520Key%2520Boundaries.csv
filePath = "/content/Eircode Routing Key Boundaries.csv"

with open(filePath, 'r') as f:
  reader = csv.reader(f)
  KeyRouteList = list(map(tuple, reader))

print(KeyRouteList)

def eircodeValidator(eircode):
    # Clean Eircode
    cleaned_eircode = re.sub(r'[^A-Za-z0-9]', '', eircode)

    # Define regex of Eircode
    eircode_pattern = r'^([AC-FHKNPRTV-Y]\d{2}|D6W)[ |\-]?[0-9AC-FHKNPRTV-Y]{4}$'

    # Check if Eircode match the Eircode regex pattern
    if re.match(eircode_pattern, cleaned_eircode):
        # Get the district of Eircode
        district = cleaned_eircode[:3]

        # Find the imformation in KeyRouteList
        for item in KeyRouteList:
            if item[0] == district:
                routing_key = cleaned_eircode[:3]
                unique_identifier = cleaned_eircode[3:]
                destination = item[1]
                print(f"Eircode = {eircode}")
                print(f"Valid Eircode, Routing Key = {routing_key}, Unique Identifier = {unique_identifier}")
                print(f"Destination = {destination}")
                print()
                return
        print(f"Eircode = {eircode}")
        print(f"Cannot find the district imformation of '{eircode}'")
        print()
    else:
        print(f"Eircode = {eircode}")
        print(f"Invalid Eircode")
        print()


eircodeValidator("V94T9PX")
eircodeValidator("V94 T9PX")
eircodeValidator("v94 T9PX")
eircodeValidator(" V94-T9PX")
eircodeValidator("V94-T9PX ")
eircodeValidator("A46-T9PX")

```
