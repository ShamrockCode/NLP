```python
import re
import time
```

```python
def printTable(table, description):
    print(f'{description}\n')
    current_row = current_col = 0
    current_row_col = re.search("^row ([0-9]+) , col ([0-9]+)$",description)
    if current_row_col:
        current_row = int(current_row_col.group(1))
        current_col= int(current_row_col.group(2))
    row_counter=0
    for row in table:
        row_counter+=1
        col_counter=0
        for col in row:
            col_counter+=1
            #print(row_counter , row, current_col, col)
            if (row_counter == current_row) and (col_counter == current_col):
                formatting = '\033[1m'+'\033[91m' #bold + red
            else: formatting = '\x1b[0m' #reset fomatting
            print(formatting + str(col).rjust(10, ' '), end=' ')  # rjust returns a 10-characters long, right justified version of the string
        print('\n\n')
    print('---------------------------------------------------------------------------------------------------------------')
```

```python
# A DP-based solution for Levenshtein distance problem
def editDistDP(x,y):
    dp = []
    dpShadow =[]
    # fill in the table with zeros
    for row in range(len(x) + 1):
        dp.append([0]* (len(y) + 1))
    for row in range(len(x) + 1):
        dpShadow.append([None]* (len(y) + 1))

    for j in range(len(y) + 1):

        dp[0][j] = j
        dpShadow[0][j] = str(j)+'←'

    for i in range(len(x) + 1):
        dp[i][0] = i
        dpShadow[i][0] = str(i)+'↑'


    for i in range(1, len(x) + 1):
        for j in range(1, len(y) + 1):

            insertion = dp[i][j-1] + 1
            deletion = dp[i-1][j] + 1

            delta = 2 if x[i-1] != y[j-1] else 0
            substitution = dp[i-1][j-1] + delta

            minValue = min(insertion, deletion, substitution)
            dp[i][j] = minValue


            finaldp = str(minValue)

            if minValue == insertion:
                finaldp += '←'
            if minValue == substitution:
                finaldp += '↖'
            if minValue == deletion:
                finaldp += '↑'
            dpShadow[i][j] = finaldp


    printTable(dpShadow,"Completed DP table after all the subproblems are solved")

    return dp[-1][-1]

time_start=time.time()
str1, str2 = "intention", "execution"
print(f'Levenshtein distance between "{str1}" and "{str2}": {editDistDP(str1, str2)}')
time_end = time.time()
time = time_end - time_start
print(f'--- Executed in: "{time}" seconds ---')
```
