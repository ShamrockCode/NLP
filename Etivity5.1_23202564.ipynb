```python
def load_words(file_path):
    with open(file_path, 'r', encoding='ISO-8859-1') as file:
        words = file.read().splitlines()
    return words

positive_words_file = '/content/positive-words.txt'
negative_words_file = '/content/negative-words.txt'

positive_words = load_words(positive_words_file)
negative_words = load_words(negative_words_file)

```

```python
def sentiment_analyzer(text):
    words = text.lower().split()

    positive_count = sum(word in positive_words for word in words)
    negative_count = sum(word in negative_words for word in words)

    confidence_score = abs(positive_count - negative_count) / len(words) if words else 0

    if positive_count > negative_count:
        return 'positive', confidence_score
    elif negative_count > positive_count:
        return 'negative', confidence_score
    else:
        return 'neutral', confidence_score

test_strings = [
    "NLP is cool",
    "NLP is cool and useful",
    "NLP is hard",
    "NLP is hard and useless",
    "NLP stands for Natural Language Processing"
]

for string in test_strings:
    sentiment, confidence = sentiment_analyzer(string)
    print(f'String = {string}\n{sentiment} sentiment\nconfidence: {confidence}\n')
```
