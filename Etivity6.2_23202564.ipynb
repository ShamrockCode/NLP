```python
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier

def naiveBayesClassifier(trainingSet, testSet):
    classifier = NaiveBayesClassifier(trainingSet)

    for i in testSet:
        print(f"document: ('{i[0]}','{i[1]}')")
        print(f"predicted class: {classifier.classify(i[0])}")

    accuracy = classifier.accuracy(testSet)
    print(f"Accuracy: {accuracy}")

trainingSet = [('London is the Capital of GB','GB'),
               ('Oxford is a city in GB','GB'),
               ('Dublin is the capital of Ireland','IE'),
               ('Limerick is a city in Ireland','IE')]
testSet = [('University of Limerick','IE'),
           ('University College Dublin','IE'),
           ('Imperial College London','GB'),
           ('University of Oxford','GB'),
           ('Ireland & GB','GB')]
naiveBayesClassifier(trainingSet, testSet)
```

```python
import pandas as pd

url = "https://storage.googleapis.com/dataset-uploader/bbc/bbc-text.csv"
data = pd.read_csv(url)

pd.set_option('display.max_colwidth', 50)

print("Sample documents:")
display(data[['category', 'text']].head(3))

print("\nNumber of document:", data.shape[0])

print("\nDocuments per category:")
display(data['category'].value_counts())
```

```python
data = data.sample(frac=1).reset_index(drop=True)

train_size = int(len(data) * 0.8)

train_set = data[:train_size]
test_set = data[train_size:]


print("Train size:", len(train_set))
print("Test size:", len(test_set))
print("\n")


print(f"Training Set Samples:\n{train_set}")
print("\n\n")
print(f"Test Set Samples:\n{test_set}")
```

```python
train_data = [tuple(x) for x in train_set[['text', 'category']].values]
test_data = [tuple(x) for x in test_set[['text', 'category']].values]

cl = NaiveBayesClassifier(train_data)

accuracy = cl.accuracy(test_data)
print("Accuracy:", accuracy)
```

10 fold cross-validation

```python
from sklearn.model_selection import KFold

def cross_validate(data, k=10):
    kf = KFold(n_splits=k, shuffle=True, random_state=1)
    accuracies = []

    for train_index, test_index in kf.split(data):
        train_data = [tuple(x) for x in data.iloc[train_index][['text', 'category']].values]
        test_data = [tuple(x) for x in data.iloc[test_index][['text', 'category']].values]

        classifier = NaiveBayesClassifier(train_data)
        accuracy = classifier.accuracy(test_data)
        accuracies.append(accuracy)
        print(f"Fold Accuracy: {accuracy}")

    print(f"Average Accuracy: {sum(accuracies) / len(accuracies)}")

cross_validate(data)
```
