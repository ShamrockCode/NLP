```python
import pandas as pd
import requests
from io import StringIO
import jellyfish
from textblob import TextBlob
```

# Task 1

```python
url = "https://norvig.com/ngrams/count_1w.txt"

response = requests.get(url)

if response.status_code == 200:
    content = response.content.decode('utf-8')
    unigrams_io = StringIO(content)
    unigrams_df = pd.read_csv(unigrams_io, sep='\t', header=None, names=['word', 'frequency'])
    display(unigrams_df.head(100))
```

```python
# Test jellyfish
print(jellyfish.damerau_levenshtein_distance('jellyfish', 'jellyfihs'))
```

```python
def nonWordSpellingCorrection(nonWord):
    candidates = []

    total_frequency = unigrams_df['frequency'].sum()

    for index, row in unigrams_df.iterrows():
        word = row['word']

        if isinstance(word, str):
            distance = jellyfish.damerau_levenshtein_distance(nonWord, word)

            if distance <= 2:
                frequency = row['frequency']
                probability = frequency / total_frequency
                candidates.append((word, frequency, probability))

    candidates_df = pd.DataFrame(candidates, columns=['word', 'frequency', 'P(word)'])

    candidates_df = candidates_df.sort_values(by='frequency', ascending=False)

    return candidates_df

usage = "acress"
candidates_df = nonWordSpellingCorrection(usage)
print(f"nonWord = {usage}")
print()
display(candidates_df)
```
